```python
# build a function that can get page title
def getTitle(url):
    html = urlopen(url)
    bs = BeautifulSoup(html.read(),'html.parser')
    title = bs.title
    return title
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import random
import datetime
```

```python
# make the process a function
def get_links(atcl_str):
    html = urlopen('http://en.wikipedia.org{}'.format(atcl_str))
    bs = BeautifulSoup(html.read(),'html.parser')
    str_set = []
    for i in bs.find('div',{'id':'bodyContent'}).find_all('a',{'href':re.compile('^(/wiki/)((?!:).)*$')}):
        str_set.append(i['href'])
        
    return str_set


# this is the method of the book, the two func is almost the same
# random.seed(datetime.datetime.now())
# def getLinks(articleUrl):
#     html = urlopen('http://en.wikipedia.org{}'.format(articleUrl))
#     bs = BeautifulSoup(html, 'html.parser')
#     return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))


# this is a loop to deep dive the web
# links = getLinks('/wiki/Kevin_Bacon')
# while len(links) > 0:
#     newArticle = links[random.randint(0, len(links)-1)].attrs['href']
#     print(newArticle)
#     links = getLinks(newArticle)
```
